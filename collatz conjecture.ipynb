### Function for calculating number of iterations on a given number number

```python
def collatz(n):
    count = 0
    while n!=1:
        count +=1 
        if n%2 == 0:
            n = n//2
            

        elif n%2 != 0:
            n = 3*n + 1
    return count
    
    
```

### Making a list to store itrations in a range

```python
def values():
    n = int(input("Enter range to find number of iterations"))
    value = [collatz(x) for x in range(1,n+1)]
    print (value)
   
```

```python
def old_values():
    n = int(input("Enter range to find number of iterations"))
    iters = []
    for i in range(1,n+1):
        iters.append(collatz(i))
        print (iters)
```
