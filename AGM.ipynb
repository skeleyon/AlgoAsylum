### Importing required libraries

```python
import numpy as np
import random
import seaborn as sns
from itertools import combinations 
import pandas as pd
import matplotlib.pyplot as plt
```

### Calculating Arithmetic and Geometric Mean

```python
def calculate(n1,n2):
    a = (n1+n2)/2       # Formula for Arithmetic mean
    g = np.sqrt(n1*n2)  # Formula for Geometric mean
    return (a, g)
```

### Calculating AGM

```python
def calculate_agm(n1,n2, a1 = None, g1 = None):
    a1 = n1
    g1 = n2
    print(a1)
    while (a1!= g1):
        agm = calculate(a1, g1)
        a1 = agm[0]
        g1 = agm[1]
        a1 = round(a1,10)       # Rounding off to 10 decimals
        g1 = round(g1,10) 
    print("The  agm is:", a1)
    return a1

```

### Taking values from the User

```python
n1 = int(input('Enter the first number:'))
n2 = int(input('Enter the second number:'))
agm = calculate_agm(n1,n2)
```

### Generating Random Pair of numbers

```python
x = int(input("Number of pairs"))
n1 = []
n2 = []

for i in range (0,x):
    y = random.randint(1,20)    # Taking random integers from 1 to 20
    z = random.randint(1,20)
    n1.append(y)
    n2.append(z)
print (n1)
print (n2)


            


    

```

### Adding AGM values in a list

```python
final_agm = []
for i in range (len(n1)):
        final_agm.append(calculate_agm(n1[i],n2[i]))
print (final_agm)
```

### Plotting AGM values

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
sns.set(style = "whitegrid")

x = (n1)
y = (n2)
z = (final_agm)

ax.set_xlabel("First Number")
ax.set_ylabel("Second Number")
ax.set_zlabel("AGM Value")

ax.scatter(x, y, z,c='purple')

plt.show()
```
